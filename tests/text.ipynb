```python
import datetime
import json

def parse_unified_market(market_data):
    """
    解析UnifiedMarket对象，提取关键信息并返回结构化字典
    修复：确保所有输出可被JSON序列化
    
    参数:
        market_data: UnifiedMarket对象，包含市场的所有原始数据
    
    返回:
        dict: 结构化的市场数据字典（所有值均可JSON序列化）
    """
    # 初始化结果字典
    parsed_data = {}
    
    # 提取基础市场信息
    parsed_data['market_id'] = market_data.market_id
    parsed_data['title'] = market_data.title
    parsed_data['url'] = market_data.url
    parsed_data['description'] = market_data.description
    parsed_data['category'] = market_data.category
    parsed_data['tags'] = market_data.tags
    parsed_data['volume_24h'] = market_data.volume_24h
    parsed_data['liquidity'] = market_data.liquidity
    parsed_data['volume'] = market_data.volume
    parsed_data['open_interest'] = market_data.open_interest
    parsed_data['image_url'] = market_data.image
    
    # 处理解析日期 - 直接转换为字符串格式（支持JSON序列化）
    parsed_data['resolution_date'] = market_data.resolution_date.strftime('%Y-%m-%d %H:%M:%S %Z')
    # 如果需要保留时间戳（数字格式），也可以用这个：
    # parsed_data['resolution_timestamp'] = int(market_data.resolution_date.timestamp())
    
    # 通用的结果项解析函数
    def parse_outcome(outcome_obj):
        if not outcome_obj:
            return None
        return {
            'outcome_id': outcome_obj.outcome_id,
            'label': outcome_obj.label,
            'price': outcome_obj.price,
            'price_change_24h': outcome_obj.price_change_24h,
            'clob_token_id': outcome_obj.metadata.get('clobTokenId'),
            'market_id': outcome_obj.market_id
        }
    
    # 解析所有结果项列表
    parsed_data['outcomes'] = [parse_outcome(oc) for oc in market_data.outcomes]
    
    # 解析yes/no/up/down特定结果项
    parsed_data['yes'] = parse_outcome(market_data.yes)
    parsed_data['no'] = parse_outcome(market_data.no)
    parsed_data['up'] = parse_outcome(market_data.up)
    parsed_data['down'] = parse_outcome(market_data.down)
    
    return parsed_data

# 自定义JSON序列化器（备用方案）
def serialize_datetime(obj):
    """
    处理JSON序列化时的datetime对象
    """
    if isinstance(obj, datetime.datetime):
        return obj.strftime('%Y-%m-%d %H:%M:%S %Z')
    raise TypeError(f"Type {type(obj)} not serializable")

# ------------------------------
# 使用示例
# ------------------------------
if __name__ == "__main__":
    # 假设你的UnifiedMarket对象在markets列表最后一位
    # parsed_result = parse_unified_market(markets[-1])
    
    # 方案1：使用修复后的解析函数（推荐）
    parsed_result = parse_unified_market(markets[1])
    print("解析后的核心数据（方案1）：")
    print(json.dumps(parsed_result, indent=4, ensure_ascii=False))
```
